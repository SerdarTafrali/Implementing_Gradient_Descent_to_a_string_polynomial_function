```python
import regex as re
import regex
```

```python
def gradient(gradient_func,x_start,learning_rate=0.01,precision=0.000001,max_iters=100):
  def read(gradient_func):
    terms = gradient_func.split('+')
    equation = [re.split('x\^?', t) for t in terms]
    eq_map = []
    for e in equation:
        try:
            coeff = int(e[0])
        except ValueError:
            coeff = 1
        try:
            power = int(e[1])
        except ValueError:
            power = 1
        except IndexError:
            power = 0
        eq_map.append((coeff, power))
    return eq_map

  def write(eq_map):
      def str_power(p):
          if p == 0:
              return ''
          elif p == 1:
              return 'x'
          else:
              return 'x^%d' % (p,)

      def str_coeff(c):
          return '' if c == 1 else str(c)
      str_terms = [(str_coeff(c) + str_power(p)) for c, p in eq_map]
      return "+".join(str_terms)

  def derivative(gradient_func):
      eq_map = read(gradient_func)
      der_map = [(p*c, p-1) for c, p in eq_map[:-1]]
      return write(der_map)

  s = derivative(gradient_func)
  s = s.replace(' ', '')
  poly = []
  for term in regex.findall(r'[+-]?\d*\w+\^?\d*', s):
      term = regex.sub(r'(\d*)([A-Za-z]\w*)(.*)', r'\1*\2\3', term)
      term = term.replace('^', '**')
      poly.append(term)

  poly = ''.join(poly)
  fx = poly

  current_x = x_start
  learning_rate = learning_rate
  precision = precision
  previous_step_size = 1 
  max_iters = max_iters
  iters = 0
  df = lambda x: eval(fx)

  while previous_step_size > precision and iters < max_iters:
    previous_x = current_x
    current_x = current_x - learning_rate * df(previous_x)
    iters = iters + 1
    print("Iteration", iters, "\nX value is", current_x)
  print("The local minimum occurs at", current_x)
```

```python
gradient("x^2+2x+1",3)
```

```python
gradient("x^2+0",3)
```
